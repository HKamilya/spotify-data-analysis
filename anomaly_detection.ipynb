```python
data.date =  pd.to_datetime(data.date, format='%Y-%m-%d')
```

```python
!pip install plotly.express
```

```python
import plotly.graph_objects as go

features = data.columns[2:]
fig = make_subplots(rows=3, cols=5)
for row in range(1, 4):
  for col in range(1, 6):
    feature = features[5*(row-1) + col - 1]
    df = pd.DataFrame(data=data, columns=['date', 'region', feature])
    df.drop(df[df['region'] != 'Россия'].index, inplace=True)
    fig.add_trace(go.Line(x=df.date, y=df[feature]), row=row, col=col)
fig.update_layout(title_text="Side By Side Subplots")
fig.show()
```

```python
df = pd.DataFrame(data=data, columns=['date', 'region', 'num_req_mortage'])
df.drop(df[df['region'] == 'Россия'].index, inplace=True)
fig = px.line(df, x='date', y="num_req_mortage", color='region', title = '')
fig.show()
```
